```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Use the raw GitHub URL
url = "https://raw.githubusercontent.com/singla-aryan/Climate-Change-Project/main/yeartempchange_data.csv"
df = pd.read_csv(url)

# Filter only the 'World' row
world_row = df[df["Country"] == "World"]

# Extract columns that are years
year_columns = [col for col in df.columns if col.isdigit()]
years = list(map(int, year_columns))
temps = world_row[year_columns].values.flatten()

# Create a clean DataFrame
clean_df = pd.DataFrame({
    "Year": years,
    "Temperature Change": temps
}).dropna()

# Plot with regression
plt.figure(figsize=(10, 6))
sns.regplot(
    x="Year",
    y="Temperature Change",
    data=clean_df,
    scatter_kws={"alpha": 0.6, "color": "blue"},
    line_kws={"color": "red"}
)
plt.title("Global Temperature Change Over Time")
plt.xlabel("Year")
plt.ylabel("Temperature Change (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlation
correlation = clean_df['Year'].corr(clean_df['Temperature Change'])
print(f"Correlation between Year and Temperature Change: {correlation:.4f}")

# Statistical analysis
slope, intercept, r_value, p_value, std_err = linregress(clean_df["Year"], clean_df["Temperature Change"])
print(f"R²: {r_value**2:.4f}")
print(f"p-value: {p_value:.4e}")
```
